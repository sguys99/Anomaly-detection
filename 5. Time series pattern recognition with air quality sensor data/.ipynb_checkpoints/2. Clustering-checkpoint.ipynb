```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings; warnings.filterwarnings('ignore')
plt.style.use('ggplot')
%matplotlib inline
```

```python
colors = ["#00798c", "#d1495b", '#edae49', '#66a182', '#e3120b', '#4a4a4a', 
          '#111D5E', '#9F5F80', '#A7C5EB', '#54E346', '#FD3A69']
sns.color_palette(colors)
```

### 1. 데이터 읽기

```python
df1 = pd.read_csv('./data/SL132001-data (11).csv')
df2 = pd.read_csv('./data/SL132001-data (12).csv')
```

```python
display(df1.head(3))
display(df1.shape)
display(df2.head(3))
display(df2.shape)
```

```python
두 데이터 병합하기 : 이때 첫번재 열은 단위 정보이므로 제외
```

```python
data = pd.concat([df1.drop(0), df2.drop(0)])
```

```python
data.head(3)
```

```python
data.shape
```

```python
data.info()
```

```python
data['timestamp'] = data['Date'] + ' ' + data['Time GMT -4']
```

```python
data['timestamp'] = pd.to_datetime(data['timestamp'], format = '%d/%m/%Y %H:%M:%S')
```

결측치 제거

```python
data = data.drop(columns=['Date', 'Time GMT -4', 'Timestamp', 'Latitude', 
                          'Longitude', 'Unnamed: 16'], axis = 1)
```

```python
data = data.set_index(keys = 'timestamp', drop = True)
```

```python
data.head()
```

각 샘플간의 인터벌 확인하기

```python
intervals = []
for i in range(1, len(data.index)):
    intervals.append((data.index[i] - data.index[i-1]).seconds)
```

```python
intervals = pd.Series(intervals)
```

```python
intervals.value_counts()
```

```python
df_cnt = pd.DataFrame()
df_cnt['intervals'] = intervals.value_counts().index.values
df_cnt['counts'] = intervals.value_counts().values
```

```python
df_cnt.head()
```

- 샘플별 인터벌의 차이가 있다.

```python
print(df_cnt['intervals'].min())
print(df_cnt['intervals'].max())
```

###  EDA

정규화 되지 않은 차트를 그려본다.

```python
fig, ax = plt.subplots(figsize = (12, 6))
data.astype('float64').plot(ax = ax, color = colors)
plt.legend(loc = 'upper right')
plt.show()
```

정규화 한후 plotting 하기

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
data = data.astype('float64')
data_norm = data.copy()
```

```python
scaler = MinMaxScaler()
data_norm[data.columns] = scaler.fit_transform(data)
```

```python
data_norm.head()
```

```python
fig, ax = plt.subplots(figsize = (12, 6))
data_norm.plot(ax = ax, color = colors, alpha = 0.8, linewidth = 0.9)
plt.legend(loc = 'upper right')
plt.show()
```

```python
cols = data_norm.columns.tolist()
```

'Temperature (External)', 'Humidity (External)', 'Ozone - Low Conc.'을 살펴본다.

```python
cols = ['Temperature (External)', 'Humidity (External)', 'Ozone - Low Conc.']
fig, ax = plt.subplots(3, 1, sharex=True, figsize = (14, 8))
for i in range(3):
    data_norm[cols[i]].plot(color = colors[i], label = cols[i], ax = ax[i])
    ax[i].set_ylabel('normalized value')
    ax[i].legend()
    ax[i].axvspan('2020-05-26 09:00:00', '2020-05-28 9:00:00', alpha=0.05, color='red')
    ax[i].axvspan('2020-05-29 18:00:00', '2020-06-01 8:00:00', alpha=0.05, color='red')
ax[0].set_title('Temp, Humid. Ozone')
plt.xlabel('timestamp')
plt.show()
```

하나씩 그려보기

```python
for col in data.columns:
    fig, ax = plt.subplots(figsize = (12, 4))
    data[col].plot(ax= ax)
    ax.set_title(col)
plt.show()
```

상관도 분석

```python
fig, ax = plt.subplots(figsize=(12, 9))
sns.heatmap(
        data.corr(), 
        cmap = sns.diverging_palette(220, 10, as_cmap = True),
        square=True, 
        cbar_kws={'shrink':.9 }, 
        ax=ax,
        annot=True, 
        linewidths=0.1,vmax=1.0, linecolor='white',
        annot_kws={'fontsize':10 })
plt.title("Pearson's Correlation Among Features")
plt.show()
```

```python
df = data_norm.iloc[:, [0, 3, 4, 7, 8, 9, 10]]
sns.pairplot(df, palette=colors)
plt.title("Scatter Matrix of Selected Features")
plt.show()
```

### Toronto weateher data와 함께 살펴보기

```python
tw = pd.read_csv('./data/weatherstats_toronto_hourly.csv')
```

```python
tw.head(3)
```

```python
tw['timestamp'] = tw['date_time_local'].str[:-4]
tw['timestamp'] = pd.to_datetime(tw['timestamp'])
```

```python
tw.head(3)
```

```python
tw = tw.set_index(keys = ['timestamp'], drop = True)
```

```python
tw.head(3)
```

```python
tw = tw[(tw.index >= '2020-05-26 09:00:00') & (tw.index <= '2020-06-09 09:00:00')]
```

```python
tw.head(3)
```

```python
tw = tw.sort_index()
```

차트 그리기

```python
fig, (ax1, ax2) = plt.subplots(2, 1, sharex = True, figsize = (14, 8))
data['Temperature (External)'].plot(ax = ax1, color = colors[0], label = 'Temperature(Exteranl)')
tw['temperature'].plot(ax = ax1, color = colors[1], label = 'Toronto Temp')
ax1.set_ylabel('temp')

data['Humidity (External)'].plot(ax = ax2, color = colors[0], label = 'Humidity(Exteranl)')
tw['relative_humidity'].plot(ax = ax2, color = colors[1], label = 'Toronto Humid')
ax2.set_ylabel('humid')

ax1.set_title('Actual and Measured Temperature and Humidity')
for ax in (ax1, ax2):
    ax.axvspan('2020-05-26 09:00:00', '2020-05-28 9:00:00', alpha=0.05, color='red')
    ax.axvspan('2020-05-29 18:00:00', '2020-06-01 8:00:00', alpha=0.05, color='red')
    ax.legend(loc='upper right')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(16, 6))
data.plot(y='Temperature (External)', use_index=True, ax=ax)
data.plot(y='Temperature (Internal)', use_index=True, ax=ax)
tw.plot(y='temperature', use_index=True, ax=ax)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(16, 6))
data.plot(y='Humidity (External)', use_index=True, ax=ax)
data.plot(y='Humidity (Internal)', use_index=True, ax=ax)
tw.plot(y='relative_humidity', use_index=True, ax=ax)
plt.show()
```

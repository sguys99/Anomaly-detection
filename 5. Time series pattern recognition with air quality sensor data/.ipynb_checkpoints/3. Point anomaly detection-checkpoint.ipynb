```python
import pickle
from sklearn.cluster import KMeans, DBSCAN
from sklearn.decomposition import PCA
from mpl_toolkits import mplot3d
```

### 1. 데이터 읽기

```python
# Load dataframes from pickle file
with open('./data/data_full.pkl', 'rb') as f:
    data, data_norm = pickle.load(f)
```

```python
def is_outlier(data, rolling_median, thresh):
    if not pd.isna(rolling_median):
        return abs(data - rolling_median) >= thresh
```

```python
data.head(3)
```

이상치 제거하기 : rolling median을 기준으로 threshold를 초과하면 해당 값을 roling 값으로 대체함

```python
data_nrm5 = data_norm.rolling(window=5, center = True).median() 
# center를 True로 하면 윈도우 중간을 기준으로 동작함
data_nc = data_norm.copy()
```

```python
s = data_nc.shape
```

```python
for col in range(s[1]):
    for row in range(s[0]):
        if is_outlier(data_nc.iloc[row, col], data_nrm5.iloc[row, col], 0.2):
            data_nc.iloc[row, col] = data_nrm5.iloc[row, col]
```

```python
data_nc.equals(data_norm)
```

```python
data_nc.shape
```

차원축소 : PCA를 사용해서 차원을 축소한다.

```python
pca = PCA(n_components=3)
data_pca3 = pca.fit_transform(data_nc)
data_pca3 = pd.DataFrame(data_pca3, columns=['PC1', 'PC2', 'PC3'])
```

```python
data_pca3.head(3)
```

### KMEANS

KMEANS 클러스터링 후 각 클러스터에 포함된 데이터들이 퍼져있는 정도를 inertia 라고 한다.
Inertia는 각 클러스터의 중심인 centroid와 각 데이터들 사이의 거리를 나타낸다.  
즉, Inertia가 낮은 그룹을 ‘좋은’그룹이라 할 수 있고, 이러한 그룹을 적게 만들수록 좋은 모델이라고 할 수 있다.  
클러스터의 수와 inertia간의 trade-off가 발생하게 된다. 우리의 목표는 inertia를 최소화 시키면서 동시에 클러스터의 수를 최소화 시키는 것이다. 

최적의 클러스터 수를 결정하는 한 가지 방법은‘elbow’ 메소드이다.

Elbow 메소드란 그래프를 사람의 팔로 간주하고, 팔꿈치 부분에 해당하는 클러스터 갯수를 선택하는 것을 말한다.  즉 inertia가 감소하는 정도가 낮아지는 지점을 찾으면 된다. 

```python
inertia = []
for k in range(1, 20):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(data_nc)
    inertia.append(kmeans.inertia_)
```

```python
fig, ax = plt.subplots(figsize = (12, 6))
ax = plt.plot(range(1, 20), inertia, color = colors[0])
ax = plt.axvline(5, linestyle = '--', color = 'red', alpha = 0.8)
plt.xlabel('# of clusters')
plt.ylabel('inertia')
plt.title('inertia vs # of clusters')
plt.show()
```

- 클러스터 수를 5로 설정한다.

```python
kmeans = KMeans(n_clusters=5, random_state=0)
labels = kmeans.fit_predict(data_nc)
data_pca3['label_kmean'] = labels
```

```python
for col in data_nc.columns:
    fig, ax, = plt.subplots(figsize = (16, 6))
    ax = plt.scatter(data_nc.index, data_nc[col], c = labels, alpha = 0.8, s = 10)
    plt.xlim([data.index[0], data.index[-1]])
    plt.title(col)
    plt.legend(['0', '1', '2', '3', '4'])
    plt.show()
```

### DBSCAN

```python
dbscan = DBSCAN(eps=0.14, min_samples=5)
labels = dbscan.fit_predict(data_norm)
data_pca3['label_DBSCAN'] = labels
```

```python
for col in data_nc.columns:
    fig, ax, = plt.subplots(figsize = (16, 6))
    ax = plt.scatter(data_nc.index, data_nc[col], c = labels, alpha = 0.8, s = 10)
    plt.xlim([data.index[0], data.index[-1]])
    plt.title(col)
    plt.legend(['0', '1', '2', '3', '4'])
    plt.show()
```

### 시각화 비교

```python
fig, ax = plt.subplots(1, 2, figsize = (12, 3))
labels = [data_pca3['label_kmean'], data_pca3['label_DBSCAN']]
titles = ['Time Series - Clustered by KMeans', 'Time Series - Clustered by DBSCAN']
for i in range(2):
    ax[i].scatter(data_nc.index, data_nc['Temperature (External)'], c = labels[i], alpha = 0.7,
                 s = 10)
    ax[i].set_xlim([data_nc.index[0], data_nc.index[-1]])
    plt.setp(ax[i].get_xticklabels(), rotation=45)
    ax[i].set_xlabel('timestamp')
    ax[i].set_ylabel('temperature (external)')
    ax[i].set_title(titles[i])
plt.show()
```

```python
# Scatter plot of first 2 PCs
fig, axs = plt.subplots(1, 2, figsize=(12, 5))
labels = [data_pca3['label_kmean'], data_pca3['label_DBSCAN']]
titles = ['PCA 2D - Clustered by KMeans', 'PCA 2D - Clustered by DBSCAN']
for i in range(2):
    axs[i].scatter(data_pca3['PC1'], data_pca3['PC2'], c=labels[i], 
                      alpha=0.5, s= 10)
    axs[i].set_xlabel('PC1')
    axs[i].set_ylabel('PC2')
    axs[i].set_title(titles[i])
plt.show()
```

```python
# Scatter plot of first 3 PCs
fig = plt.figure(figsize=(16, 6))
labels = [data_pca3['label_kmean'], data_pca3['label_DBSCAN']]
titles = ['PCA 3D - Clustered by KMeans', 'PCA 3D - Clustered by DBSCAN']

for i in range(2):
    ax = fig.add_subplot(1, 2, i+1, projection="3d")
    ax.scatter3D(data_pca3['PC1'], data_pca3['PC2'], data_pca3['PC3'], c=labels[i], 
                      alpha=0.5, s= 10)
    ax.set_xlabel('PC1')
    ax.set_ylabel('PC2')
    ax.set_zlabel('PC3')
    ax.set_title(titles[i])
    ax.view_init(30, 30)
plt.show()
```

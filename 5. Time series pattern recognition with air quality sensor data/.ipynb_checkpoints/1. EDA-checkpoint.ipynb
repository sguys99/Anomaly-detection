### 2. 데이터 준비

본 문제의 목표는 빈도수가 적은 이벤트를 조기에 검출하는 것이다.  
여기서는 4분전에 예측하는 모델을 만들어 보고자 한다.  
따라서 2분 주기인 데이터셋의 타겟 y를 2 step shift 시킬수 있을 것이다. 

하지만 여기서는 curve shift라는 방법을 사용할 것이다.  
이방법은 만약 n번째 row가 positively labeled 된 것이라면  
- n-2, n-1 번째 row도 1로 설정한다. 이로인에 분류기가 사전에 이상을 예측할 것이다.  
- n번쨰 row를 삭제한다. 고장이 발생한 시점을 분류기가 학습할 필요가 없기 때문이다. 

```python
sign = lambda x: (1, -1)[x < 0]
```

```python
def curve_shift(df, shift_by):
    '''
    This function will shift the binary labels in a dataframe.
    The curve shift will be with respect to the 1s. 
    For example, if shift is -2, the following process
    will happen: if row n is labeled as 1, then
    - Make row (n+shift_by):(n+shift_by-1) = 1.
    - Remove row n.
    i.e. the labels will be shifted up to 2 rows up.
    
    Inputs:
    df       A pandas dataframe with a binary labeled column. 
             This labeled column should be named as 'y'.
    shift_by An integer denoting the number of rows to shift.
    
    Output
    df       A dataframe with the binary labels shifted by shift.
    '''
    vector = df['y'].copy()
    for s in range(abs(shift_by)):
        tmp = vector.shift(sign(shift_by))
        tmp = tmp.fillna(0)
        vector+= tmp # 한칸식 shift 해가며 레이블을 더해간다.
                     # 단 더해져서 1이상인 값이 있을 수도 있다.
        
    labelcol = 'y'
    # Add vector to the df
    df.insert(loc=0, column=labelcol+'tmp', value=vector)
    # Remove the rows with labelcol == 1.
    df = df.drop(df[df[labelcol] == 1].index)
    # Drop labelcol and rename the tmp col as labelcol
    df = df.drop(labelcol, axis=1)
    df = df.rename(columns={labelcol+'tmp': labelcol})
    # Make the labelcol binary
    df.loc[df[labelcol] > 0, labelcol] = 1
    df['y'] = df['y'].astype('int8')

    return df
```

shift 결과 시험

```python

'''
Shift the data by 2 units, equal to 4 minutes.

Test: Testing whether the shift happened correctly.
'''
print('Before shifting')  # Positive labeled rows before shifting.
one_indexes = df.index[df['y'] == 1]
display(df.iloc[(one_indexes[0]-3):(one_indexes[0]+2), 0:5].head(n=5))

# Shift the response column y by 2 rows to do a 4-min ahead prediction.
df = curve_shift(df, shift_by = -2)

print('After shifting')  # Validating if the shift happened correctly.
display(df.iloc[(one_indexes[0]-4):(one_indexes[0]+1), 0:5].head(n=5))
```
